### Who has hit the most number of 4's

```python

four=ball_data[ball_data['batsman_runs']==4]
runs_4=four.groupby('batsman')['batsman_runs'].count().reset_index()
runs_4.columns=['Batsman','4s']
runs_4.sort_values(by='4s',ascending=False).head(1)
```

### Who has hit the most number of 6's?

```python
four=ball_data[ball_data['batsman_runs']==6]
runs_6=four.groupby('batsman')['batsman_runs'].count().reset_index()
runs_6.columns=['Batsman','6s']
runs_6.sort_values(by='6s',ascending=False).head(1)
```
